# 1. how many types of pokemons are there in the data and their values?

```python
data['Type 1'].value_counts()
```

# 2. extract the data of Fire type pokemons

```python
data.loc[data['Type 1'] == 'Fire'].head()
```

# 3. extract the data of Fire or Water or Poison and Normal type pokemons

```python
data.loc[(data['Type 1'] == 'Fire') | (data['Type 1'] == 'Water') | (data['Type 1'] == 'Poison') & (data['Type 1'] == 'Normal')]
```

# 4. Find out how many Grass type pokemons are in the data whose name ends with 'ysaur'.

```python
data.loc[(data['Type 1'] == 'Grass') & (data['Name'].str.endswith('ysaur'))]
```

#  5. display the pokemons whose defense is greater than 100.

```python
data[data['Defense'] > 100]
```

# 6. display the pokemons with HP of greater than 76 and are of water type.

```python
data[(data['HP'] > 76) & (data['Type 1'] == 'Water')].head(30)
```

# 7. find out the average Speed

```python
data['Speed'].mean()
```

# 8. display total statistics for each type 1

```python
i = 'Type 1'

for i, df  in data.groupby(['Type 1']):
    print("Type 1-",i)
    print(df['Total'].describe())
    print(df['Total'].describe().count())
```

# 9. Change legendary value to true for pokemons having total greater than 700

```python
data.loc[data['Total'] > 700, 'Legendary'] = 'superior'
data.loc[data['Legendary'] == 'superior']
```

# 10. count type 1 pokemons and use groupby func.

```python
data.groupby(['Type 1']).count()
```
